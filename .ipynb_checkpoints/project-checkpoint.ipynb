```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
sns.set(style='darkgrid', context='talk', palette='Dark2')

year_month_fmt = mdates.DateFormatter('%m/%y')


%matplotlib inline
```

```python
# import data

# Dow Jones Index
dji_data = pd.read_csv('data/^DJI.csv')

# S&P 500
gspc_data = pd.read_csv('data/^GSPC.csv')

# NASDAQ Composite
ixic_data = pd.read_csv('data/^IXIC.csv')

# Bitcoin
btc_data = pd.read_csv('data/BTC-USD.csv')

# Etherium
eth_data = pd.read_csv('data/ETH-USD.csv')

# XRP/Ripple
xrp_data = pd.read_csv('data/XRP-USD.csv')

# Gold
gold_data = pd.read_csv('data/GC=F.csv')

```

```python
print(gold_data.shape)
gold_data.head()
```

## Calculating moving averages of gold SMA

```python
gold = pd.DataFrame(gold_data[['Date', 'Close']])
gold.set_index('Date', inplace=True)
gold.head()
```

```python
gold['MA_1w'] = gold['Close'].rolling(window=7, center=True).mean()
gold['MA_3w'] = gold['Close'].rolling(window=21, center=True).mean()
gold['MA_200w'] = gold['Close'].rolling(window=1400, center=True).mean()



fig, ax = plt.subplots(figsize=(15,10))

ax.grid(True)
# ax.plot(gold['Date'], gold['Close'], label='gold')
ax.plot(gold['Close'], label='gold')
# ax.plot(gold['MA_1w'], label='1 week MA')
# ax.plot(gold['MA_3w'], label='3 week MA')
# ax.plot(gold['MA_200w'], label='200 week MA')
ax.legend(loc=2)
ax.set_ylabel('Price in $')
# ax.xaxis.set_major_formatter(year_month_fmt)
# ax.xaxis_date()
```

## EMA

```python
gold['MA_1w'] = gold['Close'].ewm(span=7, adjust=False).mean()
gold['MA_3w'] = gold['Close'].ewm(span=21, adjust=False).mean()
gold['MA_30w'] = gold['Close'].ewm(span=210, adjust=False).mean()
gold['MA_200w'] = gold['Close'].ewm(span=1400, adjust=False).mean()


plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(gold['Close'], label='gold')
plt.plot(gold['MA_1w'], label='1 week MA')
plt.plot(gold['MA_3w'], label='3 week MA')
plt.plot(gold['MA_30w'], label='30 week MA')
plt.plot(gold['MA_200w'], label='200 week MA')
plt.legend(loc=2)
ax.xaxis.set_major_formatter(year_month_fmt)
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(gold['Date'], gold['Close'], label='gold')
```

```python
gold['Change'] = np.log(gold['Close'] / gold['Close'].shift())
```

```python
plt.plot(gold['Change'])
```

## Calculating Volatility of Gold

```python
gold['volatility'] = gold['Change'].rolling(21).std().shift()
gold['volatility'].plot()
```
