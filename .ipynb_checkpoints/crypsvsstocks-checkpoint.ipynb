```python
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf

import seaborn as sns
from matplotlib import style

style.use('ggplot')
%matplotlib inline
```

```python
# import data

# Dow Jones Index
dji = pd.read_csv('data/^DJI.csv', parse_dates=True, index_col=0)

# S&P 500
gspc = pd.read_csv('data/^GSPC.csv', parse_dates=True, index_col=0)

# NASDAQ Composite
ixic = pd.read_csv('data/^IXIC.csv', parse_dates=True, index_col=0)

# Bitcoin
btc = pd.read_csv('data/BTC-USD.csv', parse_dates=True, index_col=0)

# Etherium
eth = pd.read_csv('data/ETH-USD.csv', parse_dates=True, index_col=0)

# XRP/Ripple
xrp = pd.read_csv('data/XRP-USD.csv', parse_dates=True, index_col=0)

# Gold
gold = pd.read_csv('data/GC=F.csv', parse_dates=True, index_col=0)

```

```python
print(gold.shape)
gold.head()
```

```python
gold['100ma'] = gold['Adj Close'].rolling(window=100, min_periods=0).mean()
```

```python
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1)


ax1.plot(gold.index, gold['Adj Close'])
ax1.plot(gold.index, gold['100ma'])
ax2.bar(gold.index, gold['Volume'])
```
